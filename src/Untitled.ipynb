```python
import nbformat
from nbconvert import HTMLExporter
from jinja2 import FileSystemLoader

notebook_filename = r"C:\TEMP\projects\omata\notebooks\preparatory_report3.ipynb"
output_filename = r"C:\TEMP\projects\omata\notebooks\preparatory_report3.html"

# Load notebook
with open(notebook_filename, encoding="utf-8") as f:
    nb = nbformat.read(f, as_version=4)

# HTML exporter
html_exporter = HTMLExporter()
html_exporter.exclude_input = True  # hide code

# Set up custom loader
template_path = r"C:\TEMP\projects\omata\notebooks"
html_exporter.environment.loader = FileSystemLoader(template_path)
html_exporter.template_file = "new_template.tpl"  # just the filename

body, resources = html_exporter.from_notebook_node(nb)

# Add charset and save
body = '<meta charset="utf-8">' + body
with open(output_filename, "w", encoding="utf-8") as f:
    f.write(body)

print(f"Exported HTML with margin: {output_filename}")
```
